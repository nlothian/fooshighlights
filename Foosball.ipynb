```python
import matplotlib.pyplot as plt
import cv2
```

```python
# Load the images
table = cv2.imread("images/table.jpg")
withball = cv2.imread("images/withball.jpg")
```

```python
# Show what the original images look like
%matplotlib inline
imshow(table)
```

```python
imshow(withball)
```

```python
# Convert to HSV

hsv_table = cv2.cvtColor(table, cv2.COLOR_RGB2HSV)
hsv_withball = cv2.cvtColor(withball, cv2.COLOR_RGB2HSV)
```

```python
# Now threshold both images

# These numbers don't make a lot of sense to me, but they seem to work
thresh = 160 
maxValue = 180
 
res, th_table = cv2.threshold(hsv_table, thresh, maxValue, cv2.THRESH_BINARY);

res2, th_withball = cv2.threshold(hsv_withball, thresh, maxValue, cv2.THRESH_BINARY);

# The difference between the two is the image of the ball
# I don't understand why this isn't (th_withball - th_table)??
ball_image = th_table - th_withball

%matplotlib inline
imshow(ball_image)
```
